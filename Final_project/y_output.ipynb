```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import seaborn as sb
```

```python
y_arr = np.loadtxt("output_y.txt", dtype=np.float32)
y_arr = np.abs(y_arr)
y_arr = y_arr.flatten()
y_arr.shape
```

```python
print("Mean: ", np.mean(y_arr))
plt.hist(y_arr, edgecolor='black')
plt.title("Y Value Distribution")
plt.xlabel("y-value")
plt.ylabel("number of pixels")
plt.ticklabel_format(axis="y", style="sci", scilimits=(0,0))
plt.show()
```

```python
y_map = np.reshape(y_arr, (800,800))
y_hm = sb.heatmap(y_map, cmap='Blues', vmin=0, vmax=1.6e3)
y_hm.set_title("y-value heatmap")
plt.show()
```
